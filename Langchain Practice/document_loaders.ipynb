# **Data Ingesetion - Document Loaders**

## Text Loader

```python
#Text Loader
from langchain_community.document_loaders import TextLoader
from rich import print  # for printing colorful output

loader = TextLoader("Data Transformation Simplified Expl.txt")
text_document = loader.load()
print(text_document)
```

## PDF Loader

```python
# PDF loader
from langchain_community.document_loaders import PyPDFLoader

loader = PyPDFLoader("Panaversity Cloud Native Applied Generative AI Engineer (GPTs and AI Agents).pdf")

pdf_document = loader.load() # it will load all pages and create Document of each page
print(pdf_document[:2])
print("Type of PDF Document: ", type(pdf_document))
print("Type of single document: ", type(pdf_document[0]))
```

## Webpages Loader

```python
# Web based loader
from langchain_community.document_loaders import WebBaseLoader

# we can add multiple urls
loader = WebBaseLoader(web_paths=("https://medium.com/@LawrencewleKnight/build-your-first-sql-database-agent-with-langchain-19af8064ae18",
                                  "https://ts.llamaindex.ai/modules/llamacloud"),
                       )

webpages_content = loader.load()
print(webpages_content)
```

```python
import bs4
# we can add multiple parameters
loader = WebBaseLoader(web_path=("https://ts.llamaindex.ai/modules/llamacloud"),
                       bs_kwargs= dict(parse_only=bs4.SoupStrainer(
                           # taken from the webpage
                           class_=("footer footer--dark", "language-ts codeBlockContainer_rxII theme-code-block")
                       ))
                       )

webpages_content = loader.load()
print(webpages_content)
```

## Arxive Loader

```python
# Arxiv Loader
from langchain_community.document_loaders import ArxivLoader

doc_paper = ArxivLoader(query="1706.03762", load_max_docs=1).load()  # reaserch paper number on Arxiv Website
doc_paper
```
